```python
import csv

bright_traits = ['LOGIC', 'CONCENTRATION', 'FOCUS', 'COGNITIVE SKILL', 'RETENTION POWER', 'HARD WORKING', 'STUDY HABIT', 'CONSCIOUSNESS']
dark_traits = ['SILLY MISTAKES', 'KNOWLEDGE GAP', 'IMPULSIVE', 'LEARNING GAP']

categories = {
    'A': (1, 0),
    'B': (2, 3),
    'C': (4, 6),
    'D': (7, 10)
}

csv_file = '/kaggle/input/oath-ai1/Data .csv'
student_column = 'NAME'  # Adjust this according to your CSV file

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        student = row[student_column]  # Use the correct column name for the student identifier

        bright_trait_values = []
        dark_trait_values = []

        # Calculate bright trait values
        for trait in bright_traits:
            value = float(row[trait])
            bright_trait_values.append(value)

        # Calculate dark trait values
        for trait in dark_traits:
            value = float(row[trait])
            dark_trait_values.append(value)

        # Calculate average bright traits and assign category
        bright_traits_avg = sum(bright_trait_values) / len(bright_trait_values)
        bright_category = ''
        for category, range_vals in categories.items():
            if range_vals[0] <= bright_traits_avg <= range_vals[1]:
                bright_category = category

        # Calculate average dark traits and assign category
        dark_traits_avg = sum(dark_trait_values) / len(dark_trait_values)
        dark_category = ''
        for category, range_vals in categories.items():
            if range_vals[0] <= dark_traits_avg <= range_vals[1]:
                dark_category = category

        # Print results for the student
        print(f"Student: {student}")
        print(f"Average Bright traits score: {bright_traits_avg}")
        print(f"Average Dark traits score: {dark_traits_avg}")
        print(f"Category (Bright traits): {bright_category}")
        print(f"Category (Dark traits): {dark_category}")
        print()
```

```python
import string
from collections import Counter

def count_words(text):
    # Remove punctuation marks
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Convert text to lowercase
    text = text.lower()
    
    # Split text into words
    words = text.split()
    
    # Count the frequency of each word
    word_counts = Counter(words)
    
    return word_counts

def display_top_words(word_counts, n=5):
    # Get the top n most frequent words and their frequencies
    top_words = word_counts.most_common(n)
    
    # Display the results
    print("Top", n, "most frequent words:")
    for word, count in top_words:
        print(f"{word}: {count}")

# Example usage
file_path = '/kaggle/input/oath-ai1/paragraph.txt'  # Replace with the path to your text file

with open(file_path, 'r') as file:
    text = file.read()

word_counts = count_words(text)
display_top_words(word_counts)
```

```python
import string
import chardet

def calculate_scores(text, positive_dict, negative_dict):
    # Remove punctuation marks
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Convert text to lowercase
    text = text.lower()
    
    # Split text into sentences
    sentences = text.split('.')
    
    # Remove empty sentences
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    
    # Calculate the number of sentences
    num_sentences = len(sentences)
    
    # Split text into words
    words = text.split()
    
    # Calculate the number of words
    num_words = len(words)
    
    # Calculate the Positive Score
    positive_score = sum(1 for word in words if word in positive_dict)
    
    # Calculate the Negative Score
    negative_score = sum(1 for word in words if word in negative_dict)
    
    # Calculate the Polarity Score
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    
    # Calculate the Average Sentence Length
    average_sentence_length = num_words / num_sentences
    
    return positive_score, negative_score, polarity_score, average_sentence_length

# Load positive and negative dictionaries
positive_dict_file = '/kaggle/input/oath-ai1/positive-words.txt'  # Replace with the path to your positive dictionary file
negative_dict_file = '/kaggle/input/oath-ai1/negative-words.txt'  # Replace with the path to your negative dictionary file

with open(positive_dict_file, 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    positive_encoding = result['encoding']

with open(negative_dict_file, 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    negative_encoding = result['encoding']

with open(positive_dict_file, 'r', encoding=positive_encoding) as file:
    positive_dict = file.read().splitlines()

with open(negative_dict_file, 'r', encoding=negative_encoding) as file:
    negative_dict = file.read().splitlines()

# Example usage
text_file = '/kaggle/input/oath-ai1/paragraph.txt'  # Replace with the path to your text file

with open(text_file, 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    text_encoding = result['encoding']

with open(text_file, 'r', encoding=text_encoding) as file:
    text = file.read()

positive_score, negative_score, polarity_score, average_sentence_length = calculate_scores(text, positive_dict, negative_dict)

# Display the results
print("Positive Score:", positive_score)
print("Negative Score:", negative_score)
print("Polarity Score:", polarity_score)
print("Average Sentence Length:", average_sentence_length)
```
