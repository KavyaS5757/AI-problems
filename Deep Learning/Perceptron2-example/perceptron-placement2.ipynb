```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv('/kaggle/input/perceptron-placement3/Placement_Data_Full_Class.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(data=df,x="10th %",y="Degree %",hue="Work exp")
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
